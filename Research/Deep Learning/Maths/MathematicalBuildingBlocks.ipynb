## MNIST dataset

```python
from tensorflow.keras.datasets import mnist

(train_images, train_labels)\
  , (test_images, test_labels) = mnist.load_data()
```

```python
import matplotlib.pyplot as plt
# Plot the first few digits
num_digits_to_plot = 5

for i in range(num_digits_to_plot):
    plt.subplot(1, num_digits_to_plot, i + 1)
    plt.imshow(train_images[i], cmap='gray')
    plt.title(f"Label: {train_labels[i]}")
    plt.axis('off')

plt.show()
```

```python

# Plot the first few digits with a different colormap
num_digits_to_plot = 5

for i in range(num_digits_to_plot):
    plt.subplot(1, num_digits_to_plot, i + 1)
    plt.imshow(train_images[i], cmap='magma')  # Change the colormap to 'viridis'
    plt.title(f"Label: {train_labels[i]}")
    plt.axis('off')

plt.show()
```

```python
from matplotlib.colors import LinearSegmentedColormap


# Define custom colormap using orange and blue colors
colors = [(1, 0.5, 0), (0, 0, 1)]  # Orange to blue gradient
custom_cmap = LinearSegmentedColormap.from_list("OrangeBlue", colors, N=256)

# Plot the first few digits using the custom colormap
num_digits_to_plot = 5

for i in range(num_digits_to_plot):
    plt.subplot(1, num_digits_to_plot, i + 1)
    plt.imshow(train_images[i], cmap=custom_cmap)  # Use the custom colormap
    plt.title(f"Label: {train_labels[i]}")
    plt.axis('off')

plt.show()
```
