```python
import matplotlib.pyplot as plt

X = 'Age'
Y = 'Car Purchase Amount'
plot_title = f"Scatter Plot: {X} vs {Y}"
author_name = "Alexis Tercero"
watermark_text = "alexistercero55@gmail.com"

# Set up the square figure dimensions
fig, ax = plt.subplots(figsize=(8, 8))
# Set the color of the axis spines (plot frame)
ax.spines['bottom'].set_color('#FFFFFF')
ax.spines['top'].set_color('#FFFFFF')
ax.spines['right'].set_color('#FFFFFF')
ax.spines['left'].set_color('#FFFFFF')

# Create a scatter plot
#DEPRECATED dfcsv.plot.scatter(x=X, y=Y, title=f"Scatter Plot: {X} vs {Y}")
# Create a scatter plot with a gray-blue color palette
scatter = ax.scatter(dfcsv[X], dfcsv[Y], c='#64717D', alpha=0.7, edgecolors='#88B9E3', linewidths=0.5)# Add title and labels

# Set background color
fig.patch.set_facecolor('#2C3C4A')
ax.set_facecolor('#2C3C4A')

# Add title and labels with white text color
ax.set_title(plot_title, color='#FFFFFF')
ax.set_xlabel(X, color='#FFFFFF')
ax.set_ylabel(Y, color='#FFFFFF')

# Add watermark at the bottom
ax.text(0.5, -0.1, watermark_text, transform=ax.transAxes, color='#FFFFFF', alpha=0.5, ha='center', va='center')

# Add author's name at the top
fig.text(0.5, 0.95, author_name, ha='center', va='center', fontsize=12, color='gray', alpha=0.7)

# Customize gridlines and ticks as needed
ax.grid(True, linestyle='--', alpha=0.5, color='#FFFFFF')
ax.tick_params(axis='both', which='both', length=0, color='#FFFFFF', labelcolor='#FFFFFF')  # Hide ticks

# Show the plot

plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Generate some example data
X_data = dfcsv[[X]]
y_data = dfcsv[[Y]]

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X_data, y_data)

# Make predictions on new data
X_new = np.linspace(20, 70, len(X_data)).reshape(-1, 1)
y_pred = model.predict(X_new)

# Set up the square figure dimensions
fig, ax = plt.subplots(figsize=(8, 8))
# Set the color of the axis spines (plot frame)
ax.spines['bottom'].set_color('#FFFFFF')
ax.spines['top'].set_color('#FFFFFF')
ax.spines['right'].set_color('#FFFFFF')
ax.spines['left'].set_color('#FFFFFF')

# Plot the data and the fitted line
scatter = ax.scatter(X_data, y_data, c='#64717D', alpha=0.7, edgecolors='#88B9E3', linewidths=0.5)
# Set background color
fig.patch.set_facecolor('#2C3C4A')
ax.set_facecolor('#2C3C4A')


# Add title and labels with white text color
ax.set_title(plot_title, color='#FFFFFF')
ax.set_xlabel(X, color='#FFFFFF')
ax.set_ylabel(Y, color='#FFFFFF')


# Add watermark at the bottom
ax.text(0.5, -0.1, watermark_text, transform=ax.transAxes, color='#FFFFFF', alpha=0.5, ha='center', va='center')

# Add author's name at the top
fig.text(0.5, 0.95, author_name, ha='center', va='center', fontsize=12, color='gray', alpha=0.7)

# Customize gridlines and ticks as needed
ax.grid(True, linestyle='--', alpha=0.5, color='#FFFFFF')
ax.tick_params(axis='both', which='both', length=0, color='#FFFFFF', labelcolor='#FFFFFF')  # Hide ticks
ax.plot(X_new, y_pred, color='red', label='Fitted Line')
# plt.xlabel('X')
# plt.ylabel('y')
plt.title('Simple Linear Regression with scikit-learn')
plt.legend()
plt.show()
```

```python
# Calculate residuals (errors)
residuals = y_data - y_pred

# Plot residuals
plt.scatter(X_data, residuals, color='blue', marker='o', label='Residuals')
plt.axhline(y=0, color='red', linestyle='--', label='Zero Residual Line')
plt.xlabel('Age')
plt.ylabel('Purchase residuals')
plt.title('Residuals Analysis for Simple Linear Regression')
plt.legend()
plt.show()
```
